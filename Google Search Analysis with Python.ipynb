```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import time

Trending_topics=TrendReq(hl='en-US',tz=360)
```

```python
kw_list=["cloud computing"]

Trending_topics.build_payload(
    kw_list=kw_list,
    cat=0,
    timeframe='today 12-m',
    geo='',
    gprop=''
)

time.sleep(5)
```

```python
data=Trending_topics.interest_over_time()
data=data.drop(columns=['isPartial'])
top_interest=data.sort_values(
    by="cloud computing",
    ascending=False
).head(10)
print(top_interest)
```

```python
import pandas as pd
pd.set_option('future.no_silent_downcasting', True)
```

```python
kw_list=["cloud computing"]

Trending_topics.build_payload(
    kw_list=kw_list,
    cat=0,
    timeframe='2025-10-01 2025-11-01',
    geo='',
    gprop=''
)

data=Trending_topics.interest_over_time()
data=data.drop(columns=['isPartial'])

print(data)

```

```python
keyword = "cloud computing"

data = Trending_topics.interest_over_time()
data = data.drop(columns=['isPartial'])

top_days = data.sort_values(
    by=keyword,
    ascending=False
).head(10)

print(top_days)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(data.index,data["cloud computing"])
plt.title("cloud computing search trend( nov-dec 2025)")
plt.xlabel("date")
plt.ylabel("search interest")
plt.show()
```

```python
region_data=Trending_topics.interest_by_region(
    resolution='COUNTRY',
    inc_low_vol=True
)

top_regions=(
    region_data
    .sort_values(by="cloud computing",ascending=False)
    .head(10)
)

print(top_regions)
    
```

```python
import matplotlib.pyplot as plt

top_regions["cloud computing"].plot(kind='bar', figsize=(8,4))
plt.title("Top 10 Countries by Search Interest for Cloud Computing")
plt.xlabel("Country")
plt.ylabel("Search Interest")
plt.show()
```

```python
try:
    Trending_topics.build_payload(
        kw_list=["cloud computing"],
        timeframe='today 12-m',
        geo='')
    related_queries=Trending_topics.related_queries()
    top_queries=related_queries[kw_list]['top']
    rising_queries = related_queries[kw_list]['rising']

    print("Top Related Queries:")
    print(top_queries.head())

    print("\nRising Related Queries:")
    print(rising_queries.head())

except (KeyError, IndexError, TypeError):
    print(f"No related queries found for '{kw_list}'")
    
```

```python
keywords = Trending_topics.suggestions(keyword='Cloud Computing')

df = pd.DataFrame(keywords)

df = df.drop(columns='mid')

print(df)
```
